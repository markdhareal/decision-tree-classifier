## Import Necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

## Data Preparation

```python
dataframe = pd.read_csv('../data/titanic.csv')
dataframe.head()
```

```python
dataframe = dataframe.drop(columns=['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'])
dataframe
```

```python
dataframe.isna().sum()
```

```python
dataframe.describe()
```

```python
dataframe.loc[dataframe['Age'].isna(), 'Age'] = dataframe['Age'].median()
dataframe.isna().sum()
```

```python
dataframe
```

```python
dataframe['Pclass'].unique()
```

```python
sex_encoder = LabelEncoder()
```

```python
dataframe['Sex'] = sex_encoder.fit_transform(dataframe['Sex'])
dataframe
```

## Assigning the X and y

```python
X = dataframe.drop(columns=['Survived'])
X
```

```python
y = dataframe['Survived']
y[:5]
```

## Splitting the Data into Training and Test Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

## Data Scaling

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Creation

```python
dec_tree_clf = DecisionTreeClassifier()
```

```python
dec_tree_clf.fit(X_train_scaled, y_train)
```

```python
y_pred = dec_tree_clf.predict(X_test_scaled)
y_pred[:5]
```
